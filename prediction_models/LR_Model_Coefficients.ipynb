```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from helper_functions import *
```

```python
data = pd.read_csv("./F_AD.txt", low_memory=False, header=0)
feature_list = ['DATE', 'OPEN', 'HIGH', 'LOW']
output = ['CLOSE']
data = data[feature_list+output]
data['LAG_OPEN'] = data['OPEN'].shift(1)
data['LAG_HIGH'] = data['HIGH'].shift(1)
data['LAG_LOW'] = data['LOW'].shift(1)
data['LAG_CLOSE'] = data['CLOSE'].shift(1)
data = data.dropna()
data = data.drop(['OPEN', 'HIGH', 'LOW'], axis=1)
data_train = data[data.DATE <= 20151231]
data_train = data_train[data_train.DATE >= 19900101]
data_test = data[data.DATE <= 20171231]
data_test = data_test[data_test.DATE >= 20160101]

data_train_x = data_train.drop(['DATE', 'CLOSE'], axis=1)
data_train_y = data_train.drop(['DATE', 'LAG_OPEN', 'LAG_HIGH', 'LAG_LOW', 'LAG_CLOSE'], axis=1)
data_test_x = data_test.drop(['DATE', 'CLOSE'], axis=1)
data_test_y = data_test.drop(['DATE', 'LAG_OPEN', 'LAG_HIGH', 'LAG_LOW', 'LAG_CLOSE'], axis=1)

lr = LinearRegression()
model = lr.fit(data_train_x, data_train_y)
y_pred = lr.predict(data_test_x)

coefficient = lr.coef_
intercept = lr.intercept_
rmse = rmse_ratio(data_test_y, y_pred)
mape = mean_absolute_percentage_error(data_test_y, y_pred)

result = pd.DataFrame(coefficient, columns=['LAG_OPEN', 'LAG_HIGH', 'LAG_LOW', 'LAG_CLOSE'])
result['Intercept'] = intercept
result['RMSE'] = rmse.values
result['MAPE'] = mape.values
result['Future'] = 'F_AD'
result
```

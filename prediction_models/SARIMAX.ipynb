```python
# indicator data
indicator_train = pd.read_pickle("./data_indicators_base_train.pkl")
indicator_test = pd.read_pickle("./data_indicators_base_test.pkl")
```

```python
# Create correlation matrix
corr_matrix = indicator_train.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation less than 0.4
to_drop = [column for column in upper.columns if any(upper[column] > 0.4)]

tmp = indicator_train.drop(to_drop, axis=1)

indicator_train = indicator_train[tmp.columns].values.transpose()
indicator_test = indicator_test[tmp.columns].values.transpose()
```

```python
new_features = pd.DataFrame(index=np.arange(len(test_data)+len(stack_data)), columns=np.arange(88))
new_features.columns = futures
new_features["dates"] = (list(test_data.index) + list(stack_data.index))
new_features = new_features.set_index("dates")
new_features.head()
```

```python
pointer = 0

# get close prices
x_train = train_data.iloc[:,(pointer+3)]
x_test = test_data.iloc[:,(pointer+3)]
stack_test = stack_data.iloc[:,(pointer+3)]

# take log
x_train = np.log(x_train)
x_test = np.log(x_test)
stack_test = np.log(stack_test)

period = 5 ## 5 trading days in a week

#for (fut in futures) {
  
  #pointer = pointer + 4
#}

bestSSE = np.inf
bestp = None
bestq = None
bestP = None
bestQ = None
    
def gridSearch(itemObj):
    
    currSSE = np.inf
    currp = None
    currq = None
    currP = None
    currQ = None
           
    #use gridsearch to look for optimal sarimax parameters
    for p in range(2,6):
        for q in range (2,6):
            for P in range (0, 3):
                for Q in range (0, 3):
                    if (p+q+P+Q+1+1 > 10):
                        continue
                        
                    try:
                        mod = sm.SARIMAX(itemObj, order=(p, 1, q), seasonal_order=(P, 1, Q, 5), 
                                                enforce_stationarity=False,
                                                enforce_invertibility=False)

                        results = mod.fit()
                        
                        pred = results.forecast(steps = len(x_test))
                        sse = np.sum((pred - x_test)**2)

                        #if current run of SSE is better than the best one so far, overwrite it
                        if (sse<currSSE):
                            currSSE = sse
                            currp = p
                            currq = q
                            currP = P
                            currQ = Q
                            
                    except:
                        continue
                        
        return currp, currq, currP, currQ
                    
bestp, bestq, bestP, bestQ = gridSearch(x_train)
```

```python
mod = sm.SARIMAX(x_train, order=(bestp, 1, bestq), seasonal_order=(bestP, 1, bestQ, 5),
                                                enforce_stationarity=False,
                                                enforce_invertibility=False)

results = mod.fit()
```

```python
test_dates = x_test.reset_index().iloc[:,0]
stack_dates = stack_test.reset_index().iloc[:,0]
pred = results.forecast(steps = len(x_test)+len(stack_test))

#Store predictions to use for stack model
for i in range(len(x_test)):
    new_features.at[test_dates[i], "F_AD"] = np.exp(pred[i])

ind = 0
for j in range(len(x_test),(len(x_test)+len(stack_test))):
    new_features.at[stack_dates[ind], "F_AD"] = np.exp(pred[j])
    ind = ind + 1
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from helper_functions import *
import os
```

```python
os.chdir('C:\\Users\\Cheong\\Documents\\GitHub\\ARIMAMAS\\data')

# Read data
train_data = pd.read_pickle("./data_base_train.pkl")
train_data = train_data.fillna(method='ffill', axis=0)
train_data.columns = train_data.columns.droplevel()
train_data.head()
```

```python
# Read data
test_data = pd.read_pickle("./data_base_test.pkl")
test_data = test_data.fillna(method='ffill', axis=0)
test_data.columns = test_data.columns.droplevel()
test_data.tail()
```

```python
# Read stack data
stack_data = pd.read_pickle("./data_stack_test.pkl")
stack_data = stack_data.fillna(method='ffill', axis=0)
stack_data.columns = stack_data.columns.droplevel()
stack_data.tail()
```

```python
# All 88 futures
futures = ['F_AD','F_BO','F_BP','F_C','F_CC','F_CD',
                           'F_CL','F_CT','F_DX','F_EC','F_ED','F_ES',
                           'F_FC','F_FV','F_GC','F_HG','F_HO','F_JY',
                           'F_KC','F_LB','F_LC','F_LN','F_MD','F_MP',
                           'F_NG','F_NQ','F_NR','F_O','F_OJ','F_PA',
                           'F_PL','F_RB','F_RU','F_S','F_SB','F_SF',
                           'F_SI','F_SM','F_TU','F_TY','F_US','F_W',
                           'F_XX','F_YM','F_AX','F_CA','F_DT','F_UB',
                           'F_UZ','F_GS','F_LX','F_SS','F_DL','F_ZQ',
                           'F_VX','F_AE','F_BG','F_BC','F_LU','F_DM',
                           'F_AH','F_CF','F_DZ','F_FB','F_FL','F_FM',
                           'F_FP','F_FY','F_GX','F_HP','F_LR','F_LQ',
                           'F_ND','F_NY','F_PQ','F_RR','F_RF','F_RP',
                           'F_RY','F_SH','F_SX','F_TR','F_EB','F_VF',
                           'F_VT','F_VW','F_GD','F_F']
len(futures)
```

```python
# indicator data
indicator_train = pd.read_pickle("./data_indicators_base_train.pkl")
indicator_test = pd.read_pickle("./data_indicators_base_test.pkl")
indicator_stack = pd.read_pickle("./data_indicators_stack_test.pkl")
```

```python
# Create correlation matrix
corr_matrix = indicator_train.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation less than 0.4
to_drop = [column for column in upper.columns if any(upper[column] > 0.4)]

indicator_train = indicator_train.drop(to_drop, axis=1)
indicator_test = indicator_test.drop(to_drop, axis=1)
indicator_stack = indicator_stack.drop(to_drop, axis=1)

indicator_train = indicator_train.fillna(method='bfill', axis=0)
indicator_test = indicator_test.fillna(method='bfill', axis=0)
indicator_stack = indicator_stack.fillna(method='bfill', axis=0)
```

```python
new_features = pd.DataFrame(index=np.arange(len(test_data)+len(stack_data)), columns=np.arange(88))
new_features.columns = futures
new_features["dates"] = (list(test_data.index) + list(stack_data.index))
new_features = new_features.set_index("dates")
new_features.head()
```

```python
from sklearn.preprocessing import StandardScaler
import lightgbm as lgb

scaler = StandardScaler()

count = 0

# Build separate LGBM models for each individual future
for fut in futures:
    
    # Lag features by 1 day
    x_train = train_data.shift(1).iloc[:,count:(count+4)]
    # combine with indicators
    combined_train = pd.concat([x_train, indicator_train], axis=1)
    combined_train = combined_train.dropna(axis=0)
    x_train = combined_train.iloc[:,0:4]
    x_train_exog = combined_train.iloc[:,4:]
    y_train = train_data.iloc[:,(count+3)]
    y_train = y_train.dropna(axis=0)
    y_train = y_train.drop(y_train.index[0])
    y_train = y_train.reset_index()
    train_dates = y_train.iloc[:,0]
    y_train = y_train.iloc[:,1]

    x_test = test_data.shift(1).iloc[:,count:(count+4)]
    combined_test = pd.concat([x_test, indicator_test], axis=1)
    combined_test = combined_test.dropna(axis=0)
    x_test = x_test.dropna(axis=0)
    x_test = combined_test.iloc[:,0:4]
    x_test_exog = combined_test.iloc[:,4:]
    y_test = test_data.iloc[:,(count+3)]
    y_test = y_test.dropna(axis=0)
    y_test = y_test.drop(y_test.index[0])
    y_test = y_test.reset_index()
    test_dates = y_test.iloc[:,0]
    y_test = y_test.iloc[:,1]

    x_stack = stack_data.shift(1).iloc[:,count:(count+4)]
    combined_stack = pd.concat([x_stack, indicator_stack], axis=1)
    combined_stack = combined_stack.dropna(axis=0)
    x_stack = x_stack.dropna(axis=0)
    x_stack = combined_stack.iloc[:,0:4]
    x_stack_exog = combined_stack.iloc[:,4:]
    y_stack = stack_data.iloc[:,(count+3)]
    y_stack = y_stack.dropna(axis=0)
    y_stack = y_stack.drop(y_stack.index[0])
    y_stack = y_stack.reset_index()
    stack_dates = y_stack.iloc[:,0]
    y_stack = y_stack.iloc[:,1]

    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    x_stack = scaler.transform(x_stack)

    d_train = lgb.Dataset(x_train, label=y_train)

    #Set model params
    params = {}
    params['learning_rate'] = 0.5          
    params['boosting_type'] = 'gbdt'
    params['objective'] = 'regression'
    params['metric'] = 'mse'
    params['sub_feature'] = 0.8
    params['num_leaves'] = 20               
    params['min_data'] = 20
    params['max_depth'] = 20

    lgbmodel = lgb.train(params, d_train, 100)

    #Predictions
    y_pred = lgbmodel.predict(x_test)
    stack_pred = lgbmodel.predict(x_stack)

    #Store predictions to use for stack model
    for i in range(len(y_pred)):
        new_features.at[test_dates[i], fut] = y_pred[i]
    for j in range(len(stack_pred)):
        new_features.at[stack_dates[j], fut] = stack_pred[j]

    count = count + 4
    
    print('Mean Absolute Error for ', fut, ': ', mean_absolute_percentage_error(y_test, y_pred))  
    print('Root Mean Squared Error for ', fut, ': ', rmse_ratio(y_test, y_pred))
```
